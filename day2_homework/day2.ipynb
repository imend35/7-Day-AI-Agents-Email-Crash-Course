STEP 1 : DATA GENERATION

```python
import requests

urls = [
    "https://raw.githubusercontent.com/ageron/handson-ml3/main/README.md",
    "https://raw.githubusercontent.com/pytorch/pytorch/main/README.md",
    "https://raw.githubusercontent.com/openai/openai-cookbook/main/README.md",
    "https://raw.githubusercontent.com/langchain-ai/langchain/main/README.md",
    "https://raw.githubusercontent.com/fastapi/fastapi/master/README.md"
]

docs = []

for url in urls:
    res = requests.get(url)

    docs.append({
        "title": url.split("/")[-1],
        "description": "GitHub markdown doc",
        "content": res.text
    })

print(len(docs))
print(docs[0]["content"][:3000])
```

STEP 2 — Simple Chunking (Sliding Window)

```python
def sliding_window(text, size=2000, step=1000):
    chunks = []
    for i in range(0, len(text), step):
        chunk = text[i:i+size]
        chunks.append(chunk)
        if i + size >= len(text):
            break
    return chunks
```

```python
simple_chunks = []

for doc in docs:
    chunks = sliding_window(doc["content"])

    for i, ch in enumerate(chunks):
        simple_chunks.append({
            "title": doc["title"],
            "chunk_id": i,
            "chunk": ch
        })

print(len(simple_chunks))
print(simple_chunks[0]["chunk"][:3000])
```

STEP 3 — Paragraph Chunking

```python
import re

def split_paragraphs(text):
    return re.split(r"\n\s*\n", text.strip())
```

```python
paragraph_chunks = []

for doc in docs:
    paragraphs = split_paragraphs(doc["content"])

    for i, p in enumerate(paragraphs):
        if len(p.strip()) > 100:  # çok kısa olanları ele
            paragraph_chunks.append({
                "title": doc["title"],
                "chunk_id": i,
                "chunk": p
            })

print(len(paragraph_chunks))
print(paragraph_chunks[0]["chunk"][:3000])
```

STEP 4 — Section Chunking

```python
def split_sections(text):
    sections = re.split(r"\n## ", text)
    return sections
```

```python
section_chunks = []

for doc in docs:
    sections = split_sections(doc["content"])

    for i, sec in enumerate(sections):
        if len(sec.strip()) > 200:
            section_chunks.append({
                "title": doc["title"],
                "chunk_id": i,
                "chunk": sec
            })

print(len(section_chunks))
print(section_chunks[0]["chunk"][:3000])
```

STEP 5 — Comparison

```python
print("Simple:", len(simple_chunks))
print("Paragraph:", len(paragraph_chunks))
print("Section:", len(section_chunks))
```

STEP 6 — Analysis Article

I applied three chunking strategies on real-world GitHub markdown documentation: 
simple sliding window chunking, paragraph-based chunking, and section-based chunking.

Simple chunking was easy to implement and ensured uniform chunk sizes, but it often split sentences and disrupted semantic meaning.

Paragraph-based chunking improved readability, but resulted in inconsistent chunk sizes and sometimes fragmented context in technical documentation.

Section-based chunking performed best because markdown documents naturally contain structured headers. This approach preserved semantic coherence and produced self-contained chunks, making it more suitable for retrieval-based applications.
